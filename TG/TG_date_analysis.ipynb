```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import date
date = pd.read_csv('date.csv', skiprows=31, encoding="GBK")
date.info()
```

```python
# draw the plot
fig, ax = plt.subplots(3, 1)

x1 = date['##Temp./°C']
x2 = date['Time/min']
y1 = date['Mass/%']  # TG 
y2 = date['DTA/(uV/mg)']  # DTA
y3 = date['DTG/(%/min)']  # DTG

# TG
ax[0].plot(x1, y1, 'b-')

# DTA
ax2 = ax[1]
ax2.plot(x2, y2, 'r-')

# DTG
ax3 = ax[2]
ax3.plot(x2, y3, 'g-')
```
