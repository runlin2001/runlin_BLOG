```python
import numpy as np
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(2000)
x = np.arange(5)
y = np.random.randint(-5, 5, 5)

fig,axes = plt.subplots(ncols=2)
v_bars = axes[0].bar(x,y,color='red')  # 画一个bar图
h_bars = axes[1].barh(x,y,color='red')  # 画一个bar，且水平展示的图：barh


axes[0].axhline(0, color='grey', linewidth=2)  # 加一条横着的线
axes[1].axvline(0, color='grey', linewidth=2)  # 加一条竖直的线
plt.show()
```

```python
fig, ax = plt.subplots()
v_bars = ax.bar(x,y,color='lightblue')
for bar, height in zip(v_bars, y):
    if height < 0:
        bar.set(edgecolor = 'darkred', color = 'green', linewidth = 3)
```
