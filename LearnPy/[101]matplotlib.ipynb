```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
index= np.arange(0,100,10),
columns = ['A','B','C','D'])
df.head()
```

The cumsum function in NumPy calculates the cumulative sum of the elements along a given axis. In the example you provided, np.random.randn(10, 4) generates a 10x4 array of random numbers drawn from a normal distribution. The cumsum function will then calculate the cumulative sum of these elements along axis 0, resulting in a 10x4 array where each element is the sum of all the elements up to and including that element along axis 0.

```python
df.plot()
```
