Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])  # 其他参数自动设置
```

加label

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.xlabel('xlabel')
plt.ylabel('ylabel')
```

改标签大小

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.xlabel('xlabel',fontsize=16)
plt.ylabel('ylabel')
```

不同的线条

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],':')
plt.xlabel('xlabel',fontsize=16)
plt.ylabel('ylabel',fontsize=16)
```

换颜色

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25], '-.', color='y')
plt.xlabel('xlabel',fontsize=16)
plt.ylabel('ylabel',fontsize=16)
```

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'y-.')  # 也可以这样去写这个参数
plt.xlabel('xlabel',fontsize=16)
plt.ylabel('ylabel',fontsize=16)
```

```python
pu_numpy = np.arange(0,10,0.5)
plt.plot(pu_numpy, pu_numpy, 'r--')
plt.plot(pu_numpy, pu_numpy**2, 'b.-')
plt.plot(pu_numpy, pu_numpy**3, 'go')
```

也可以写进去

```python
pu_numpy = np.arange(0,10,0.5)
plt.plot(pu_numpy, pu_numpy, 'r--',
pu_numpy, pu_numpy**2, 'b.-',
pu_numpy, pu_numpy**3, 'go')
```

指定线条宽度

```python
x = np.linspace(-10,10)
y = np.sin(x)
plt.plot(x,y, linewidth=3.0)
```

marker点

```python
plt.plot(x,y,color='r',linestyle=':',marker='o')
```

```python
plt.plot(x,y,color='r',linestyle=':',marker='o',markerfacecolor='b',markersize=10)
```

也可以先画图，再设置参数

```python
line = plt.plot(x,y)
plt.setp(line,color='r',linewidth=2.0,alpha=0.2)
```

子图

```python
plt.subplot(211)  # 211表示的含义是一会要画的图是两行一列，最后一个1表示的是子图当中的第一个图
plt.plot(x,y,color='r')

plt.subplot(212)  # 最后一个2表示的是子图当中的第二个图
plt.plot(x,y,color='b')
```

```python
plt.subplot(121)  # 211表示的含义是一会要画的图是一行两列，最后一个1表示的是子图当中的第一个图
plt.plot(x,y,color='r')

plt.subplot(122)  # 最后一个2表示的是子图当中的第二个图
plt.plot(x,y,color='b')
```

```python
plt.subplot(221)
plt.plot(x,y,color='r')

plt.subplot(224)
plt.plot(x,y,color='b')
```

```python
plt.subplot(321)
plt.plot(x,y,color='r')

plt.subplot(324)
plt.plot(x,y,color='b')
```

加标签、注释

```python
plt.plot(x,y,color='r',linestyle=':',marker='o')
plt.xlabel('x...')
plt.ylabel('y....')
plt.title('wo')
plt.text(0,0,'sdwadsd')  # 前两个是坐标
plt.grid(True)  # 加格子
plt.annotate('sdawds', xy=(-5,0), xytext=(-2,0.4), arrowprops= dict(facecolor='blue',shrink=0.05,headwidth=10,headlength=10))  # 画箭头注释
```
