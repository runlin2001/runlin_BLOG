## Duck typing
很多时候我们并不在乎一个object的类型，而是更在乎它的用法，方法。我们把它称为duck typing。 If it walks like a duck and quacks like a duck, then it is a duck. 例如，你可以验证一个对象如果使用了*迭代器协议*_(iterator protocol)_，那它一定是迭代(iterable)的。对于很多object来说它包含一个magic method：`_iter_`，但是使用`iter`函数是一个更好的、独立的方法。

````python
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError:  # 不可遍历
        return False
````

````python
isiterable('a string')
````

````python
isiterable([1, 2, 3])
````

````python
isiterable(5)
````

编写接受多种输入类型的函数时，这个功能很常用。可以检查对象是否为一个列表或者numpy数组，如果不是就把它转化为列表。
```
if not isinstance(x, list) and isiterable(x):
    x = list(x)
```

## Import
从同个路径下的另一个文件连接到some_module.py中定义的变量和函数，可以这样做：

````python
import some_module
result = some_module.f(5)
pi = some_module.PI
print(result)
print(pi)
````

导入的方式有很多，还比如：

````python
from some_module import f, g, PI
result = g(5, PI)
result
````

通过使用as关键字，可以对导入内容给予不同的变量名：

````python
import some_module as  sm
from some_module import PI as pi, g as gf

r1 = sm.f(pi)
r2 = gf(6, pi)
print(r1, r2)
````
