```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import date
date = pd.read_csv('xrd.txt',header=None)  # heaher=None means no columns in original table, so that it will not make the 1st raw be the columns.
date.head()
```

```python
# adjust date for future ploting
date = date[0].str.split(expand=True)  # split columns
date = date.rename(columns={0:'Two theta', 1:'Intensity'})  # rename columns
date.head()
```

```python
date = date.astype(float)  #  这里查找最大值的时候发现了问题，当下的这个DateFrame是object类型的，所以这里要将数据类型改成float，才能正常画图
date.max()
```

```python
date.min()
```

```python
x = date['Two theta']
y = date['Intensity']
plt.plot(x, y, linewidth=0.8, color='black')  
plt.axis([5,60, 5,500])  # set the scales of x and y
plt.xlabel('Two theta')
plt.ylabel('Relative Intensity')
plt.title('XRD date')
plt.yticks([])  # remove ticks on y axis
plt.minorticks_on()  # add minorticks
```

```python
plt.subplot(211)
x = date['Two theta']
y = date['Intensity']
plt.plot(x, y, linewidth=0.8, color='black')  
plt.axis([5,60, 5,500])  # set the scales of x and y
plt.ylabel('Relative Intensity')
plt.yticks([])  # remove ticks on y axis
plt.minorticks_on()  # add minorticks

plt.subplot(212)
x = date['Two theta']
y = date['Intensity']
plt.plot(x, y, linewidth=0.8, color='r')  
plt.axis([5,60, 5,500])  # set the scales of x and y
plt.xlabel('Two theta')
plt.ylabel('Relative Intensity')
plt.yticks([])  # remove ticks on y axis
plt.minorticks_on()  # add minorticks

plt.subplot(211)
plt.title('XRD date')
```
